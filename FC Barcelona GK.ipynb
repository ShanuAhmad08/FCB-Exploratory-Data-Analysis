```python
#Importing the pandas libraries
import pandas as pd
raw_df = pd.read_csv(r"C:\Users\anand\Desktop\FCB Stats\FC Barcelona GK stats.csv")
raw_df
```

```python
#Cleaning the dataframe
df = raw_df.drop(raw_df[raw_df['Player'] == 'Squad Total'].index)
df
```

```python
#Importing the matplotlib libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Comparing the number of matches played
x = df['Matches Played']
y = df['Player']
plt.xlabel('Players')
plt.ylabel('Matches Played')
plt.title('Number of mathces played by each goalkeeper')
plt.bar(y,x, width = 0.1, color = 'khaki')
plt.show()
```

```python
#The ratio of goals conceeded per match - The less the better
df['Goals Against per match'] = df['Goals Against']/df['Matches Played']
#print(df['Goals Against per match'])
x = df['Player']
y = df['Goals Against per match']
plt.xlabel('Players')
plt.ylabel('Goals conceeded per match')
plt.title('Goals conceeded per match')
plt.bar(x,y, width = 0.1)
plt.show()
```

```python
#Importing the numpy libraries
import numpy as np
```

```python
#Comparing the shots on target faced and saves made per match
df['Shots on Target per match'] = df['Shots on Target Against']/df['Matches Played']
#print(df['Shots on Target per match'])
df['Saves per match'] = df['Saves']/df['Matches Played']
#print(df['Saves per match'])
x = df['Player']
y = df['Shots on Target per match']
z = df['Saves per match']
xpos = np.arange(len(x))
plt.xticks(xpos, x)
plt.xlabel('Players')
plt.ylabel('Shots and Saves')
plt.bar(xpos, y, width = 0.1, label = 'Shots on Target per match', color = 'mediumturquoise')
plt.bar(xpos, z, width = 0.1, label = 'Saves per match', color = 'coral')
plt.legend(fontsize = 8, loc = 'upper left', framealpha = 0)
plt.title('Approximate of Goalkeeping abilities')
plt.show()
```

```python
#Knowing how many points out of available did the goalkeeper help in achieve
df['Points gained out of available'] = (df['W']*3 + df['D']*1 + df['L']*0)/(df['W']*3 + df['D']*3 + df['L']*3)
x = df['Player']
y = df['Points gained out of available']
plt.xlabel('Players')
plt.ylabel('Points percentage')
plt.bar(x,y, width = 0.1, color = 'darkorange')
plt.title('Points gained out of available')
plt.show()
```

```python
#Comparing the clean sheet stats of both the goalkeepers - The more the better
df['% of clean sheets'] = df['CS']/(df['W'] + df['D'] + df['L'])
x = df['Player']
y = df['% of clean sheets']
plt.xlabel('Players')
plt.ylabel('Clean Sheet %')
plt.bar(x,y, width = 0.1, color = 'salmon')
plt.title('Percentage of matches with clean sheets')
plt.show()
```
