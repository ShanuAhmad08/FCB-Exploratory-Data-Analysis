```python
#Importing pandas libraries
import pandas as pd
#Importing the raw database
raw_df = pd.read_csv(r'C:\Users\anand\Desktop\FCB Stats\FC Barcelona Passing Type stats.csv')
raw_df
```

```python
#Removing the goalkeepers because not involved much in outfield play
filtered_df = raw_df[raw_df['Pos'] != 'GK']
filtered_df.drop(filtered_df[filtered_df['Player'] == 'Squad Total'].index, inplace = True)
filtered_df.drop(filtered_df[filtered_df['Player'] == 'Opponent Total'].index, inplace = True)
filtered_df.drop(filtered_df[filtered_df['90s'] < 10].index, inplace = True)
filtered_df
```

```python
#Separating the defenders
defenders_df = filtered_df[filtered_df['Pos'].str.startswith('DF')]
defenders_df
```

```python
#Importing the matplotlib libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Pass Completion % - The more the better
defenders_df['Pass Completion %'] = defenders_df['Passes Completed']*100/defenders_df['Passes Attempted']
y = defenders_df['Pass Completion %']
x = defenders_df['Player']
plt.barh(x,y, height = 0.5)
plt.title('Pass Completion %')
plt.show()
```

```python
#Passes blocked per 90 - The less the better
defenders_df['Passes Blocked per match'] = defenders_df['Passes Blocked']/defenders_df['90s']
x = defenders_df['Player']
y = defenders_df['Passes Blocked per match']
plt.barh(x,y, height = 0.5)
plt.title('Number of passes made that were blocked')
plt.show()
```

```python
#Separating the midfielders
midfielders_df = filtered_df[filtered_df['Pos'].str.startswith('MF')]
midfielders_df
```

```python
#Pass completion % - The more the better
midfielders_df['Pass completion %'] = midfielders_df['Passes Completed']*100/midfielders_df['Passes Attempted']
x = midfielders_df['Player']
y = midfielders_df['Pass completion %']
plt.ylabel('Players')
plt.xlabel('Pass Completion %')
plt.barh(x,y, height = 0.5, color = 'turquoise')
plt.title('Percentage of passes completed')
plt.show()
```

```python
#Indicating which players played the most through balls
plt.pie(midfielders_df['Through Balls'], labels = midfielders_df['Player'], autopct = '%.1f%%')
plt.title('Number of through balls played')
plt.show()
```

```python
#Indicating which players took the most freekicks
plt.pie(midfielders_df['FK'], labels = midfielders_df['Player'], autopct = '%.1f%%', explode = [0,0,0,0.5,0,0])
plt.title('Number of freekicks attempted')
plt.show()
```

```python
#Showing which player switched the ball the most
plt.pie(midfielders_df['Switches'], labels = midfielders_df['Player'], autopct = '%.1f%%')
plt.title('Number of switches made')
plt.show()
```

```python
#Estimating the number of crosses
plt.pie(midfielders_df['Crosses'], labels = midfielders_df['Player'], autopct = '%.1f%%')
plt.title('Number of crosses made')
plt.show()
```

```python
#Separating the attackers
attackers_df = filtered_df[filtered_df['Pos'].str.startswith('FW')]
attackers_df
```

```python
#Pass completion % - The more the better
attackers_df['Pass completion %'] = attackers_df['Passes Completed']*100/attackers_df['Passes Attempted']
x = attackers_df['Pass completion %']
y = attackers_df['Player']
plt.xlabel('Pass Completion %')
plt.ylabel('Players')
plt.barh(y,x, height = 0.5)
plt.title('% of successful passes played')
plt.show()
```

```python
#Through balls
plt.pie(attackers_df['Through Balls'], labels = attackers_df['Player'], autopct = '%.1f%%')
plt.title('Number of through balls attempted')
plt.show()
```

```python
#Crosses
plt.pie(attackers_df['Crosses'], labels = attackers_df['Player'], autopct = '%.1f%%')
plt.title('Number of crosses made')
plt.show()
```

```python
#Switches
plt.pie(attackers_df['Switches'], labels = attackers_df['Player'], autopct = '%.1f%%')
plt.title('Number of switches made')
plt.show()
```
