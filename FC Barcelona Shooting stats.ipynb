```python
import pandas as pd
raw_df = pd.read_csv(r'C:\Users\anand\Desktop\FCB Stats\FC Barcelona Shooting stats.csv')
raw_df
```

```python
#Removing the players with 0 goals for a better comparision
filtered_df = raw_df.drop(raw_df[raw_df['Goals'] == 0].index)
filtered_df.reset_index(inplace = True)
del filtered_df['index']
filtered_df
```

```python
#Importing matpllotlib libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Goal conversion rate - The more the better
filtered_df['Goal Conversion Rate'] = filtered_df['Goals']*100/filtered_df['xG']
x = filtered_df['Goal Conversion Rate']
y = filtered_df['Player']
plt.barh(y,x)
plt.show()
#The players with Goal Conversion Rate more than 100 have scored more than what was expected of them and vice-versa
```

```python
#Making a df to get penalty stats
penalty_df = filtered_df[filtered_df['PK attempted'] != 0]
penalty_df
```

```python
#Penalty conversion rate - The more the better
penalty_df['Penalty Conversion Rate'] = penalty_df['PK scored']*100/penalty_df['PK attempted']
x = penalty_df['Penalty Conversion Rate']
y = penalty_df['Player']
plt.barh(y,x, height = 0.3, color = 'red')
plt.show()
```

```python
#Calculating the number of freekicks attempted
FK_df = filtered_df[filtered_df['Shots from FK'] != 0]
FK_df
```

```python
plt.pie(FK_df['Shots from FK'], labels = FK_df['Player'])
plt.show()
```

```python
#Shot accuracy %
filtered_df['Shot accuracy %'] = filtered_df['Shots on Target']*100/filtered_df['Shots']
filtered_df['Shot accuracy %'] = filtered_df['Shot accuracy %'].astype(int)
filtered_df
```
